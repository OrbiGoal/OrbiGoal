```python
# Function to generate URL
def generate_logo(squad_name):
    base_url = "https://raw.githubusercontent.com/your-username/football-logos/main/logos/"
    filename = squad_to_filename.get(squad_name, "").replace(" ", "%20") + ".png"
    return f"{base_url}{filename}" if filename else None
```

```python
# Map URL to logo and map full name to squad
team_names = team_2022[["Squad", "Country"]]
team_names = team_names.copy()
team_names.loc[:, 'Logo'] = team_names['Squad'].apply(generate_logo)
squad_to_full_name = {k: v.split('/')[-1].replace('%20', ' ').replace('.png', '') for k, v in squad_to_filename.items()} # Get full name of squad
team_names.loc[:, 'full_name'] = team_names['Squad'].map(squad_to_full_name)
team_names.head()
```
